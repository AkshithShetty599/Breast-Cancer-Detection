```python
import pandas as pd

df = pd.read_csv(r"/content/drive/MyDrive/Colab Notebooks/Breast Cancer Detection/Data/data.csv")
```

```python
df.info()
```

```python
df = df.drop(['id','Unnamed: 32'],axis=1)
```

```python
df.columns
```

```python
df['diagnosis'].value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].map({'B':0,'M':1})
```

```python
df['diagnosis'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

features = [
            'radius_mean',
            'texture_mean',
            'perimeter_mean',
            'area_mean',
            'smoothness_mean',
            'compactness_mean',
            'concavity_mean',
            'concave points_mean',
            'symmetry_mean',
            'fractal_dimension_mean',
            'radius_se',
            'texture_se',
            'perimeter_se',
            'area_se',
            'smoothness_se',
            'compactness_se',
            'concavity_se',
            'concave points_se',
            'symmetry_se',
            'fractal_dimension_se',
            'radius_worst',
            'texture_worst',
            'perimeter_worst',
            'area_worst',
            'smoothness_worst',
            'compactness_worst',
            'concavity_worst',
            'concave points_worst',
            'symmetry_worst',
            'fractal_dimension_worst'
            ]
target = 'diagnosis'

X = df[features]
y = df[target]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2023)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gbm_untuned = GradientBoostingClassifier(random_state=2023)

gbm_untuned.fit(X_train,y_train)

print(f"Training Accuracy : {gbm_untuned.score(X_train,y_train)}")
print(f"Test Accuracy : {gbm_untuned.score(X_test,y_test)}")
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,gbm_untuned.predict(X_test))
```

```python
from sklearn.metrics import classification_report

print("\nClassification Report:\n", classification_report(y_test, gbm_untuned.predict(X_test)))
```

```python
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=2023)

params = {
    'n_estimators': [100, 300, 500],  # smaller range
    'learning_rate': [0.01, 0.1],     # most impactful values
    'max_features': [None, "sqrt"],
    'subsample': [0.9, 1.0],          # fewer options
    'max_depth': [3, 5, 7],           # avoid deeper trees
    'min_samples_leaf': [2, 5]        # reduced options
}

grid = RandomizedSearchCV(
    gb,
    params,
    n_iter=30,
    scoring="recall",
    n_jobs=-1
)

grid.fit(X_train, y_train)

grid.best_params_
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

# Best parameters from RandomizedSearchCV
best_params = {
    'subsample': [0.8,0.9],
    'n_estimators': [300],
    'min_samples_leaf': [1,2],
    'max_features': [None],
    'max_depth': [1,2,3],
    'learning_rate': [0.1]
}

# Define the model
gb = GradientBoostingClassifier(random_state=2023)

# GridSearchCV for confirmation + refit
grid = GridSearchCV(
    gb,
    best_params,
    scoring='recall',
    cv=5,
    n_jobs=-1
)

# Fit the grid search
grid.fit(X_train, y_train)

# Best model (already trained)
best_model = grid.best_estimator_

# Optional: Evaluate
y_pred = best_model.predict(X_test)

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import pickle
with open("model.pkl", "wb") as f:
    pickle.dump(best_model, f)
```
